Here, I add a column to the dataframe called churn_rate. Churn rate is a measure of the number of employees moving out of a company within a specific period of time. In this case, it's months. The calculation of churn rate is denoted by the number of users lost divided by the sum of total users at the start of a time period and the number of users added. 

```python
df['churn_rate'] = (df['outflow']/(df['inflow'] + df['count']))
df.head(20)
```

I create a new dataframe that looks only at row observations with sector equal to Information Technology, and group those values by job category and find the mean churn rate for each job category.

```python
df_churn_mean = df.loc[df['sector']=='Information Technology'].groupby('job_category', as_index=False)['churn_rate'].mean().sort_values(by='churn_rate')
df_churn_mean = df_churn_mean.set_index('job_category')
df_churn_mean
```

```python
df_churn_median = df.loc[df['sector']=='Information Technology'].groupby('job_category', as_index=False)['churn_rate'].median().sort_values(by='churn_rate')
df_churn_median = df_churn_median.set_index('job_category')
df_churn_median
#The same is done in this case but with median churn rate instead of mean 
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1)

df_churn_mean.plot.barh(ax=ax[0], figsize=(12,15))
df_churn_median.plot.barh(ax=ax[1])


ax[0].set_title('Mean Churn Rates for Job Categories in IT', fontsize=18)
ax[0].set_ylabel('Job Categories', fontsize=14)
ax[0].set_xlabel('Mean Churn Rates', fontsize=14)
ax[0].legend({'Churn Rate'}, bbox_to_anchor=(1, 1))

ax[1].set_title('Median Churn Rates for Job Categories in IT', fontsize=18)
ax[1].set_xlabel('Median Churn Rates', fontsize=14)
ax[1].set_ylabel('Job Categories', fontsize=14)
ax[1].legend({'Churn Rate'}, bbox_to_anchor=(1, 1))

#Creation of the graph using matplotlib fig and ax objects. 
#Plotted two horizontal bar charts one on top of the other and set axis titles, labels, and legend placement
#Plotting the graphs side by side had too many overlapping labels, so I opted to 2 rows 1 col instead
```

Creates a new dataframe that takes the top 10 companies in the Information Technology industry with the highest mean counts.

Creates a new dataframe that looks at the most current month of September 2020 and groups by companies within the IT industry and looks for their mean churn rate for each company.

```python
df_current_churn = df.loc[(df['sector']=='Information Technology') & (df['month']=='2020-09')].groupby('company', as_index=False)['churn_rate'].mean().sort_values(by='churn_rate', ascending=True)
df_current_churn = df_current_churn.set_index('company')
df_current_churn
```

Same thing but uses month values that are exactly one year before the previous month value

```python
df_past_churn = df.loc[(df['sector']=='Information Technology') & (df['month']=='2019-09')].groupby('company', as_index=False)['churn_rate'].mean().sort_values(by='churn_rate')
df_past_churn = df_past_churn.set_index('company')
df_past_churn
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1)

#Creates two bar charts one on top of the other for current and past churn rates
df_current_churn.plot.barh(ax=ax[0], figsize=(12,15))
df_past_churn.plot.barh(ax=ax[1])

ax[0].set_title('Mean Churn Rates for Companies in 2020-09', fontsize=18)
ax[0].set_ylabel('Companies', fontsize=14)
ax[0].set_xlabel('Mean Churn Rates in 2020-09', fontsize=14)
ax[0].legend({'Churn Rate'}, bbox_to_anchor=(1,1))

ax[1].set_title('Mean Churn Rates for Companies in 2019-09', fontsize=18)
ax[1].set_ylabel('Companies', fontsize=14)
ax[1].set_xlabel('Mean Churn Rates in 2019-09', fontsize=14)
ax[1].legend({'Churn Rate'}, bbox_to_anchor=(1,1))
```

These graphs show the mean churn rates for all companies within the IT industry in 2019 and 2020. In 2020, the highest churn rate went to Atos SE, a multinational IT and consulting service company. Whereas in 2019, this spot went to Capgemini. The churn rates for 2020 were overall higher, with Atos SE having a churn rate of 1.75%, and Capgemini having a churn rate of 1.6%. This could be due to the impact of covid-19 on the job market. Lots of employees lost their jobs because of the pandemic, which a likely reason why churn rate was higher in 2020 when compared to 2019. In the lower end of the spectrum, companies such as Apple and HP consistently exhibit low churn rates. This might be because many of the jobs at Apple and HP are high-skilled and the company cannot afford to have such a high employee turnover rate. However, possibly due to covid-19, the churn rate for Apple and HP in 2020 are significantly higher than their 2019 counterparts, with around 0.8% and 0.4% respectively.
